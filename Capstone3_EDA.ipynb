```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
%matplotlib inline
pd.pandas.set_option('display.max_columns',None)
```

```python
os.chdir('/Users/kofiassabil/Downloads')
```

```python
df = pd.read_csv('dataWrangling.csv')
```

```python
#median house price of all the years
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs YearSold")
```

```python
#plotting sale price histogram
sns.distplot(df['SalePrice'], color='r', bins=50, hist_kws={'alpha': 0.2})
```

```python
#shows outliers because histogram is skewed to the right
#looks like outliers are any house after the price of $450,000
```

```python
#edit df to only show houses less than or equal to $450000
df = df[df['SalePrice']<=450000]
```

```python
#correlation between variables and plotting it
corr = df.corr(method = 'pearson')
corr.round(2).style.background_gradient(cmap='coolwarm')
```

```python
plt.figure(figsize=(30,15))
sns.heatmap(corr, vmax=1, annot=True, square=True)
plt.show()
```
